# MAP & REPLACE
Used mostly with Series (single column) to transform values using a function, dictionary, or Series. three function can use in mapping.

Dictionary â†’ when you want direct value-to-value mapping.

Function â†’ when you want a condition or transformation logic. apply nu kodbhudu apply means series and data frame iradu kododhu

Series â†’ when mapping comes from another Series.

```python
l1 = pd.DataFrame({
    'Names':['a','b','c','d']
})
l1['Names'].map({'a':'A','b':'B','d':'D'})
```

```python
l2 = pd.DataFrame({
    'Marks':[25,24,22,21]
})
l2['Marks'].map({25:84,24:88,22:87,21:82})
```

```python
l2 = pd.DataFrame({
    'Marks':[2,8,5,4]
})
l2['Marks'].map(lambda x: x * 2)
```

```python
#apply()
df = pd.DataFrame({
    'math': [10, 20, 30],
    'sci':  [40, 50, 60]
})

# row-wise (axis=1) -> math + sci
df['total'] = df.apply(lambda x: x['math'] + x['sci'], axis=1)

print(df)
```

```python
d1=pd.DataFrame({
    'keys':[101,102,103,104]
})
l3 =pd.Series({
    101:'pass',
    102:'fail',
    103:'paas'
})
d1['keys'].map(l3)
```

# GroupBy

```python
data = {
    'City': ['Bengaluru', 'Mysuru', 'Bengaluru', 'Hubli', 'Mysuru', 'Hubli'],
    'Name': ['Asha', 'Ravi', 'Meena', 'Kiran', 'Anu', 'Raju'],
    'Sales': [250, 300, 200, 150, 400, 350]
}

df = pd.DataFrame(data)
print(df)
```

```python
df.groupby('City')["Sales"].sum()
```

```python
df.groupby('City')["Sales"].min()
```

```python
df.groupby('City')["Sales"].max()
```

```python
df.groupby('City')["Sales"].mean()
```

```python
df.groupby('City')["Sales"].count()
```

```python
df.groupby('City')["Sales"].agg(['min','max','count','mean','sum'])
```
